```python
### Prepare data for modeling ###
```

```python
X = df.drop(columns=['readmitted','readmitted_30d','encounter_id','patient_nbr'])
y = df['readmitted_30d']

X=pd.get_dummies(X,drop_first=True)
```

```python
### Train-Test Split ###

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
### Handle Class Imbalance - Class Weights ###

from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression(class_weight = 'balanced', max_iter=1000)
model_lr.fit(X_train, y_train)
```

```python
### Logistic Regression ###

model_lr = LogisticRegression(class_weight = 'balanced', max_iter=1000)
model_lr.fit(X_train, y_train)
```

```python
### Decision Treee ###

from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier(class_weight='balanced', max_depth=5)
model_dt.fit(X_train, y_train)
```

```python
### Random Forest ###

from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(
    n_estimators=100,
    class_weight='balanced',
    random_state=42
)

model_rf.fit(X_train,y_train)
```

```python
### Recall, Precision, F1-Score, AUC-ROC ###
```

```python
from sklearn.metrics import classification_report, roc_auc_score

def evaluate(model):
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:,1]

    print(classification_report(y_test,y_pred))
    print('AUC:', roc_auc_score(y_test,y_prob))
```

```python
evaluate(model_lr)
evaluate(model_dt)
evaluate(model_rf)
```
